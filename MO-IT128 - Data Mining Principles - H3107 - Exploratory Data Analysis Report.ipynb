# Step 1: Import Libraries and Load Data
In this step, we import necessary Python libraries such as pandas, numpy, and matplotlib for data manipulation and visualization. We also load the three datasets into pandas DataFrames.

# Step 7: Analyze Correlations

This step focuses on calculating and visualizing the correlation matrix to identify relationships between numerical variables in the merged dataset. We address potential issues such as chained assignment, missing values, and non-numeric columns to ensure accurate correlation analysis.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customer_feedback = pd.read_csv("Customer_Feedback_Data.csv")
transaction_data = pd.read_csv("Transaction_Data.csv")

# Step 1: Handle missing values and data types in 'Satisfaction_Score'
# Correctly handle missing values in 'Satisfaction_Score' in customer_feedback
customer_feedback['Satisfaction_Score'] = pd.to_numeric(customer_feedback['Satisfaction_Score'], errors='coerce')
customer_feedback['Satisfaction_Score'] = customer_feedback['Satisfaction_Score'].fillna(customer_feedback['Satisfaction_Score'].median())

# Ensure 'Satisfaction_Score' is of integer type
customer_feedback['Satisfaction_Score'] = customer_feedback['Satisfaction_Score'].astype(int)

# Step 2: Merge Transaction Data with Customer Feedback Data on 'Customer_ID'
merged_data = transaction_data.merge(customer_feedback, on='Customer_ID', how='left')

# Display the merged data
print("Merged Data:")
display(merged_data.head())

# Step 3: Identify non-numeric columns
non_numeric_cols = merged_data.select_dtypes(exclude=['number']).columns
print("Non-numeric columns:", non_numeric_cols)

# Step 4: Convert 'Transaction_Date' to datetime objects
merged_data['Transaction_Date'] = pd.to_datetime(merged_data['Transaction_Date'])

# Calculate the time difference from a reference date (e.g., the earliest date in the dataset)
reference_date = merged_data['Transaction_Date'].min()
merged_data['Days_Since_Start'] = (merged_data['Transaction_Date'] - reference_date).dt.days

# Impute missing values in 'Days_Since_Start' with 0
merged_data['Days_Since_Start'] = merged_data['Days_Since_Start'].fillna(0)

# Ensure 'Days_Since_Start' is numeric
merged_data['Days_Since_Start'] = pd.to_numeric(merged_data['Days_Since_Start'], errors='coerce').fillna(0)

# Step 5: Select only numeric columns
numeric_data = merged_data.select_dtypes(include=[np.number])

# Display the numeric data
print("\nNumeric Data:")
display(numeric_data.head())

# Step 6: Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Step 7: Visualize the correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numeric Columns Only)')
plt.show()
```

# Step 8: Identify Outliers

This step involves identifying outliers in the numerical variables of the merged dataset. We'll focus on 'Satisfaction_Score' and use boxplots and the IQR method to detect potential outliers.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customer_feedback = pd.read_csv("Customer_Feedback_Data.csv")
transaction_data = pd.read_csv("Transaction_Data.csv")

# Clean Satisfaction_Score: Convert to numeric and fill missing values with the median
customer_feedback['Satisfaction_Score'] = pd.to_numeric(customer_feedback['Satisfaction_Score'], errors='coerce')
customer_feedback['Satisfaction_Score'] = customer_feedback['Satisfaction_Score'].fillna(customer_feedback['Satisfaction_Score'].median())
customer_feedback['Satisfaction_Score'] = customer_feedback['Satisfaction_Score'].astype(int)

# Merge Transaction Data with Customer Feedback Data on 'Customer_ID'
merged_data = transaction_data.merge(customer_feedback, on='Customer_ID', how='left')

# Display the merged data
print("Merged Data:")
display(merged_data.head())

# --- Identify Outliers in 'Satisfaction_Score' ---

# 1. Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=merged_data['Satisfaction_Score'])
plt.title('Boxplot of Satisfaction Score')
plt.ylabel('Satisfaction Score')
plt.show()

# 2. IQR Method
Q1 = merged_data['Satisfaction_Score'].quantile(0.25)
Q3 = merged_data['Satisfaction_Score'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = merged_data[(merged_data['Satisfaction_Score'] < lower_bound) | (merged_data['Satisfaction_Score'] > upper_bound)]

print("\nNumber of outliers:", len(outliers))
print("\nOutliers Data:")
display(outliers[['Customer_ID', 'Satisfaction_Score']])

# Optional: Remove outliers (Be cautious when removing data)
# merged_data = merged_data[(merged_data['Satisfaction_Score'] >= lower_bound) & (merged_data['Satisfaction_Score'] <= upper_bound)]
# print("\nData after removing outliers:", merged_data.shape)
```

# Step 9: Generate Insights

Based on the exploratory data analysis, we derive insights about customer satisfaction, transaction patterns, and potential areas for improvement. These insights can inform business decisions and strategies.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customer_feedback = pd.read_csv("Customer_Feedback_Data.csv")
transaction_data = pd.read_csv("Transaction_Data.csv")

# Data Cleaning for 'Satisfaction_Score' in customer_feedback
customer_feedback['Satisfaction_Score'] = pd.to_numeric(customer_feedback['Satisfaction_Score'], errors='coerce')
customer_feedback['Satisfaction_Score'] = customer_feedback['Satisfaction_Score'].fillna(customer_feedback['Satisfaction_Score'].median())
customer_feedback['Satisfaction_Score'] = customer_feedback['Satisfaction_Score'].astype(int)

# Merge Transaction Data with Customer Feedback Data on 'Customer_ID'
merged_data = transaction_data.merge(customer_feedback, on='Customer_ID', how='left')

# Convert 'Transaction_Date' to datetime objects
merged_data['Transaction_Date'] = pd.to_datetime(merged_data['Transaction_Date'])

# Calculate time since the first transaction
reference_date = merged_data['Transaction_Date'].min()
merged_data['Days_Since_Start'] = (merged_data['Transaction_Date'] - reference_date).dt.days

# Fill missing values in 'Days_Since_Start' with 0
merged_data['Days_Since_Start'] = merged_data['Days_Since_Start'].fillna(0)

# Ensure 'Days_Since_Start' is numeric
merged_data['Days_Since_Start'] = pd.to_numeric(merged_data['Days_Since_Start'], errors='coerce').fillna(0)

#--- Generating Insights ---

# 1. Overall Customer Satisfaction
avg_satisfaction = merged_data['Satisfaction_Score'].mean()
print(f"Overall Average Satisfaction Score: {avg_satisfaction:.2f}\n")

# 2. Satisfaction Trends Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Transaction_Date', y='Satisfaction_Score', data=merged_data.sort_values('Transaction_Date'))
plt.title('Satisfaction Score Trend Over Time')
plt.xlabel('Transaction Date')
plt.ylabel('Satisfaction Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Correlation Analysis
correlation = merged_data['Transaction_Amount'].corr(merged_data['Satisfaction_Score'])
print(f"Correlation between Transaction Amount and Satisfaction Score: {correlation:.2f}\n")

# 4. Distribution of Satisfaction Scores
plt.figure(figsize=(8, 6))
sns.histplot(merged_data['Satisfaction_Score'], kde=True, bins=10)
plt.title('Distribution of Satisfaction Scores')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.show()

# 5. Low Satisfaction Analysis
low_satisfaction = merged_data[merged_data['Satisfaction_Score'] <= 4]
print("\nCustomers with Low Satisfaction (Score <= 4):")
display(low_satisfaction[['Customer_ID', 'Satisfaction_Score', 'Transaction_Amount', 'Transaction_Date']])

# 6. High Satisfaction Analysis
high_satisfaction = merged_data[merged_data['Satisfaction_Score'] >= 9]
print("\nCustomers with High Satisfaction (Score >= 9):")
display(high_satisfaction[['Customer_ID', 'Satisfaction_Score', 'Transaction_Amount', 'Transaction_Date']])

# 7. Feedback Analysis
print("\nSample Feedback Comments:")
print(customer_feedback['Feedback_Comments'].sample(10))
```
